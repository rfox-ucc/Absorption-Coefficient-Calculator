```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
import os
from decimal import Decimal, getcontext
```

```python


# Get the list of all files and directories
path = "/content/drive/MyDrive/Autocoupling/Coupling Run 11-08-23 Raw"
dir_list = os.listdir(path)
dir_list.sort()
# prints all files
print(dir_list)
```

```python
num_skip=5
for item in dir_list:
  if not (".csv" in item):
    dir_list.remove(item)
wavelengths = np.empty([])
gamma = np.empty([])
gamma_err = np.empty([])
for file in dir_list:
  arr = np.loadtxt(f"{path}/{file}",
                  delimiter=",", dtype=str, skiprows = num_skip)
  interval = float(arr[:,1][1])-float(arr[:,1][0])
  range = float(arr[:,1][-1])-float(arr[:,1][0])
  #print(range,interval)
  dut = float(arr[:,0][0])-interval*num_skip
  wavelengths = np.append(wavelengths,dut)
  outer_out_list = []
  for inner_list in arr:
      innet_out_list = []
      for string in inner_list:
          innet_out_list.append(float(string))
      outer_out_list.append(innet_out_list)

  arr = np.array(outer_out_list)

  def lin_fit_func(x, a, b):
      return a*x + b
  arr1 = scipy.optimize.curve_fit(lin_fit_func, arr[:,0],arr[:,1], p0=[5E-8, 0.009])
  xx = arr[:,0]
  yy = -arr1[0][0]*(xx-(dut+range/2))
  yy2=arr[:,1]+yy
  hh = scipy.fft.rfft(yy2)
  maxfreq = (np.abs(hh[1:]).argmax())

  freq = scipy.fft.rfftfreq(len(arr),interval)
  maxfreq1 = freq[maxfreq+1]

  spacing = max([((1/maxfreq1)/interval)-10,1])
  midline = np.mean(yy2)
  hipeaks, _ = scipy.signal.find_peaks(yy2,distance=spacing, height = midline)
  lopeaks, _ = scipy.signal.find_peaks(-yy2,distance=spacing, height = -1*midline)
  fig = plt.figure(figsize=(10,6))
  ax1 = fig.add_subplot(111)
  ax1.set_title("Fabry Perot Fringes")
  ax1.set_xlabel('Wavelength (nm)')
  ax1.set_ylabel('Optical Power (uW)')

  ax1.plot(arr[:,0],arr[:,1], c='r', label='the data')


  ax1.plot(xx,yy2, c='g', label='the result')
  ax1.plot(xx[hipeaks],yy2[hipeaks], "x")
  ax1.plot(xx[lopeaks],yy2[lopeaks], "x")
  plt.show()
  Gamma = []
  for i in hipeaks:
    difference_array1 = np.absolute(xx[lopeaks]-xx[i])
    index1 = np.argpartition(difference_array1,1)
    print(f"Nearest values to the high peak {xx[i]} are : {xx[lopeaks[index1[0]]]},{xx[lopeaks[index1[1]]]}")
    k1 = (yy2[i]-(yy2[lopeaks[index1[0]]]))/(yy2[i]+(yy2[lopeaks[index1[0]]]))
    k2 = (yy2[i]-(yy2[lopeaks[index1[1]]]))/(yy2[i]+(yy2[lopeaks[index1[1]]]))
    Rprime1 = (1/k1)*(1-np.sqrt(1-k1**2))
    Rprime2 = (1/k2)*(1-np.sqrt(1-k2**2))

    print(k1,k2)
    Gamma.append((np.log(0.34/Rprime1))/0.4)
    Gamma.append((np.log(0.34/Rprime2))/0.4)
    print(Gamma[-1])
  gamma_mean = sum(Gamma)/len(Gamma)
  gamma = np.append(gamma,gamma_mean)
  gamma_err = np.append(gamma_err, np.std(np.array(Gamma), dtype=np.float64))
  print(f"{dut}nm, Calculated Loss in dB: Gamma  {gamma_mean}")
  print("/n")
  print(f"{dut}nm, Standard Deviation in dB: Gamma  {gamma_err[-1]}")
fig1 = plt.figure()
ax2 = fig1.add_subplot(111)
wavelengths = np.delete(wavelengths,0)
gamma = np.delete(gamma,0)
gamma_err = np.delete(gamma_err,0)
print(gamma_err)
ax2.plot(wavelengths,gamma,c='r')
ax2.errorbar(wavelengths, gamma, gamma_err, fmt='o', linewidth=2, capsize=6)
ax2.set_title("Material Loss vs Wavelength")
ax2.set_xlabel('Wavelength (nm)')
ax2.set_ylabel('Loss (dB)')

plt.show()
```
